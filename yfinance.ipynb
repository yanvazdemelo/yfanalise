```python
# Recouperação de datas. Sistma SIMPLESISSIMO. Muito a melhorar.
class DateOperations:
    def __init__(self, date):
        self.date = date

    def pastdays(self, days):
        pastdate = self.date - timedelta(days=days)
        pastdate = pastdate.strftime("%Y-%m-%d")
        pastdate = datetime.strptime(pastdate, "%Y-%m-%d")
        return pastdate

    def pastweeks(self, weeks):
        days = weeks * 7
        return self.pastdays(days)

    def pastmonths(self, months):
        days = months * 30
        return self.pastdays(days)

    def pastyears(self, years):
        days = years * 365
        return self.pastdays(days)

    def pastperiod(self, period, value):
        if period == "d":
            return self.pastdays(value)
        elif period == "w":
            return self.pastweeks(value)
        elif period == "m":
            return self.pastmonths(value)
        elif period == "y":
            return self.pastyears(value)
        else:
            raise ValueError("Period should be 'days = d', 'weeks = w', 'months = m', or 'years = y'")

def dateop(period, time):
    date_operation = DateOperations(datetime.now())
    date = date_operation.pastperiod(period, time)
    return date

def dateop_cut(lastdate, period, time):
    date_operation = DateOperations(lastdate)
    date = date_operation.pastperiod(period, time)
    return date
```

```python
petr4 = sqltodf("petr4.sa")
petr4['date'] = pd.to_datetime(petr4['date'])



def regression(df):
    # Converter datas para números inteiros (número de dias desde uma data de referência)
    df['Date_Integer'] = (df['date'] - pd.Timestamp('1970-01-01')).dt.days

    # Definir X e Y
    X = df[['Date_Integer']]  # Data como números inteiros
    Y = df['open']  # Valores

    # Criação do modelo de regressão linear
    model = LinearRegression()

    # Treinamento do modelo nos dados de treinamento
    model.fit(X, Y)

    # Previsões nos dados de teste
    predictions = model.predict(X)
    
    b2 = model.coef_[0]
    b1 = model.intercept_
    # Avaliação do desempenho do modelo (opcional)
    mse = mean_squared_error(Y, predictions)
    print(f'''Regressão 
        {b1} = b1
        {b2} = b2
        {mse} = mse
        ----''')
    return b1, b2, mse
# Chamar a função com o DataFrame 'petr4'
regression(petr4)

```

```python
# Variância média entre o High e o Low de uma série de dados. Amplitude do movimento através do dia

def highlowvar(df):
    hl_diff = []
    for index, row in df.iterrows():
        high = row['high']
        low = row['low']
        diff = high - low
        hl_diff.append(diff)
    mean = np.mean(hl_diff)
    return mean
    
```

```python
def sqldiff(tabela, data1, period, num):
    df = sqltodf(tabela)
    df = cutdf_time(df, data1, period, num)
    df.sort_values(by='date', inplace=True)
    valor_mais_antigo = df['open'].iloc[0]
    valor_mais_recente = df['open'].iloc[-1]
    diff = valor_mais_recente - valor_mais_antigo
    return diff
test = sqldiff('petr4_sa', todaydt, "d", 3)
print(test)
```

```python
import paramiko

def enviar_arquivo_via_sftp(local_file, remote_dir, remote_filename, host, username, password):
    try:
        # Criar uma conexão SSH
        transport = paramiko.Transport((host, 22))
        transport.connect(username=username, password=password)

        # Criar uma sessão SFTP
        sftp = paramiko.SFTPClient.from_transport(transport)

        # Enviar o arquivo local para o servidor remoto
        sftp.put(local_file, f'{remote_dir}/{remote_filename}')

        # Fechar a conexão SFTP
        sftp.close()

        print(f'O arquivo "{local_file}" foi enviado com sucesso para o diretório "{remote_dir}" no servidor remoto.')

    except Exception as e:
        print(f'Ocorreu um erro: {e}')

# Definir informações de conexão
host = '192.168.15.57'
username = 'yan'
password = '121212'

# Definir o caminho local do arquivo e o nome do arquivo remoto
local_file = 'yfhistoric.db'
remote_dir = '/home/yan'
remote_filename = 'yfhistoric.db'

# Chamar a função para enviar o arquivo
enviar_arquivo_via_sftp(local_file, remote_dir, remote_filename, host, username, password)
```
