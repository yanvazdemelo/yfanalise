```python
import sqlite3

# Conectar ao banco de dados SQLite
conn = sqlite3.connect('yfhistoric.db')
cursor = conn.cursor()

# Obter todos os nomes de tabela do banco de dados
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
table_names = cursor.fetchall()

# Iterar sobre os nomes das tabelas e aplicar as substituições desejadas
for old_name in table_names:
    new_name = old_name[0].replace("=", "_com_").replace(".", "_").replace("^", "STOCK_")
    # Verificar se o novo nome já existe
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table' AND name=?", (new_name,))
    existing_table = cursor.fetchone()
    if existing_table:
        print(f"Tabela '{new_name}' já existe. Ignorando renomeação.")
    else:
        # Renomear a tabela
        cursor.execute(f"ALTER TABLE '{old_name[0]}' RENAME TO '{new_name}'")

# Confirmar as alterações
conn.commit()

# Fechar a conexão com o banco de dados
conn.close()
```
